```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

df=pd.read_csv("Life Expectancy Data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=df["Year"], y=df["Life expectancy "], marker='o', color='b')

plt.xlabel("Year", fontsize=12)
plt.ylabel("Life Expectancy", fontsize=12)
plt.title("Life Expectancy Over Time", fontsize=14)
plt.grid(True)
plt.show()
```
